```python
# Importando as Bibliotecas

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando a Variavel X

x = np.linspace(0,5,11)
```

```python
# Criando funções Y(X) e Z(X)

y = np.power(x-2,2.0)
z = np.exp(x/2.0)
```

```python
# Criando Grid

plt.plot(x,y,label='$y(x)=(x-2)^2$')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x)')
plt.grid(linestyle = '-' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```

```python
# Alterando a Cor do fundo

plt.plot(x,y,label='$y(x)=(x-2)^2$')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('y(x)')
plt.grid(linestyle = '--' , which = 'major' , color = 'black')
plt.legend()
plt.show()
```
