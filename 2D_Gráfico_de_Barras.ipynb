```python
 # Importando as bibliotecas e classes do projeto

 import numpy as np # trabalha com objetos vetor e matriz multidimensional
 import matplotlib.pyplot as plt # gerar os graficos de visualização
```

```python
# Criando os campos Descritivos (Labels)

labels = ["A1","A2","A3","A4","A5"]
```

```python
# Criando os valores dos campos (Metricas do Grafico)
valores = [10, 34, 56, 78, 12]
```

```python
# Visualizando o Grafico com o objeto plt (Matplotlib)
plt.bar(labels,valores)
```

```python
# Visualizando o Grafico com o objeto plt (Matplotlib)
plt.barh(labels,valores)
```

```python
# Atribuindo novos aos eixos do grafico (x, y)
plt.bar(labels,valores,color='green')
plt.xlabel('Codigo do Produto')
plt.ylabel('Valor do Produto')
```

```python
# Colando os valores existentes em cada tipo como rotulo de dados e mudando a cor do Gráfico
plt.bar(labels,valores,color='red')
plt.xlabel('Codigo do Produto')
plt.ylabel('Valor do Produto')

for x,y in zip(labels,valores):
  plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom')
```

```python
plt.bar(labels,valores,color='purple')
plt.xlabel('Codigo do Produto')
plt.ylabel('Valor do Produto')

for x,y in zip(labels,valores):
  plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom')
```

```python
# Dois Graficos de Barras em uma Visualização
```

```python
# Criando um novo Vetor chamado valores2.
valores2 = [8, 30, 60, 21, 30]
```

```python
from ssl import PROTOCOL_TLSv1_1
# Plotar p Grafico com os dois graficos e elementos de valores juntos
largura = 0.3
pos1 = np.arange(len(valores))
pos2 = [x + largura for x in pos1]

plt.bar(pos1,valores, color='black', width=largura, label='Barra 1')
plt.bar(pos2, valores2, color='purple', width=largura, label='Barra 2')
plt.legend()
```

```python
plt.bar(pos1,valores,label = "Barra 1,",color= 'black')
plt.bar(pos1,valores2,label = "Barra 2",color= 'pink',bottom = valores)
```

```python
plt.barh(pos1,valores,label = "Barra 1",color= 'pink')
plt.barh(pos1,valores2,label = "Barra 2",color= 'black',left = valores)
```
