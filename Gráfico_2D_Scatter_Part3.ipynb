```python
# Importando as Bibliotecas 

import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Criando Variavel X 

X = np.linspace(0,5,11)
```

```python
# Criando Funções Y(X) e Z(X)

Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
```

```python
# Criando Gráfico de Y em função de X 

plt.scatter(X,Y)
plt.show()
```

```python
# Alterando tamanho do ponto 

plt.scatter(X,Y,s=14)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.show()
```

```python
# Inserindo a Legenda e mais um Scatter

plt.scatter(X,Y,s=14,label='Y(X)')
plt.scatter(X,Z,s=40,label='Z(X)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```

```python
# Alterando a Densidade das cores

plt.scatter(X,Y,s=10,label='Y(X)',alpha=0.8)
plt.scatter(X,Z,s=40,label='Z(X)',alpha=0.4)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.legend()
plt.show()
```
