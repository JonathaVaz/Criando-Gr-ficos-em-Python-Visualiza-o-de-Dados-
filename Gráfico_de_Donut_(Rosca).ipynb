```python
# Importando as Bibliotecas 

import numpy as np # Vetor e Matriz Multidimencional
import matplotlib.pyplot as plt # Criar Gráficos e objetos visualização de dados
```

```python
# Criando as labels 

labels = ["Grupo 1","Grupo 2","Grupo 3","Grupo 4","Grupo 5"]
```

```python
# Criando lista com valores

valores = [10,45,32,56,23]
```

```python
# Criando lista com cores

cores = ['purple','cyan','silver','magenta','green']
```

```python
# Definindo raio

raio = 0.7
```

```python
# Criando Gráfico de Barras 

plt.pie(valores,labels=labels,colors=cores)
circulo = plt.Circle((0,0),raio,color='white')
p = plt.gcf()
p.gca().add_artist(circulo)
plt.show()
```

```python
# Adicionando porcentagem 

plt.pie(valores,labels=labels,colors=cores,autopct='%1.1f%%')
circulo = plt.Circle((0,0),raio,color='white')
p = plt.gcf()
p.gca().add_artist(circulo)
plt.show()
```
