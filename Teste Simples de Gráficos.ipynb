```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
genero = ["Rock", "Hip Hop", "Pop", "Jazz", "Reggae"]
quantidade_msc = [350, 570, 233, 190, 290]

plt.style.use("dark_background")
colors = sns.color_palette('plasma')
```

```python
fig = plt.figure(figsize=(11,7))
plt.title("\nPopularidade/Quantidade dos Gêneros Musicais do DataSet\n", fontsize=20)
plt.bar(genero, quantidade_msc, color=(colors), edgecolor='white')
```

```python
fig = plt.figure(figsize=(11,7))
plt.title("Popularidade dos Gêneros Musicais do DataSet")
plt.barh(genero, quantidade_msc, color=(colors), edgecolor='white')
```

```python
fig = plt.figure(figsize=(11,7))
plt.title('Quantidade de Musicas dos Gêneros Musicais em Porcentagem')
plt.pie(quantidade_msc, labels=genero, autopct='%1.1f%%', colors=(colors))
plt.show()
```

```python
plt.style.use('dark_background')
colors = sns.color_palette('cool')


fig = plt.figure(figsize=(11,7))
plt.title('Quantidade de Musicas dos Gêneros Musicais em Porcentagem', fontsize=14, style='italic', fontweight='bold')
plt.pie(quantidade_msc, labels=genero, autopct='%1.1f%%', colors=(colors))
plt.show()
```
